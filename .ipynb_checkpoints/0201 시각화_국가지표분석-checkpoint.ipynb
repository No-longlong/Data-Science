★ 주어진 데이터에는 여러 가지 지표가 column에 포함되어 있습니다.


기대 수명: 'Life expectancy at birth- years'

인터넷 사용자 비율: 'Internet users percentage of population 2014'

숲 면적 비율: 'Forest area percentage of total land area 2012'

탄소 배출 증가율: 'Carbon dioxide emissionsAverage annual growth'

```python
%matplotlib inline
import matplotlib.pylab as plt
import pandas as pd
```

```python
df = pd.read_csv('data/world_indexes.csv')
```

```python
life = df['Life expectancy at birth- years']
internet = df['Internet users percentage of population 2014']
forest = df['Forest area percentage of total land area 2012']
carbon = df['Carbon dioxide emissionsAverage annual growth']
```

```python
plt.scatter(df['Life expectancy at birth- years'], df['Internet users percentage of population 2014'])
```

```python
plt.scatter(life, forest)
```

```python
plt.scatter(life, carbon)
```

```python
plt.scatter(internet, forest)
```

```python
plt.scatter(internet, carbon)
```

```python
plt.scatter(forest, carbon)
```

#### 기대수명과 인터넷 사용자 비율의 상관성이 가장 높게 나타난다.
