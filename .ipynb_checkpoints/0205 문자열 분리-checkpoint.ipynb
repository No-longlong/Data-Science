# 10. 직업 탐구하기
이번에는 여자 비율이 높은 직업과, 남자 비율이 높은 직업이 무엇인지 궁금한데요.


groupby 문법을 사용해서 '여성 비율'이 높은 순으로 직업을 나열해 보세요.


DataFrame이 아닌 Series로, 'gender'에 대한 값만 아래와 같이 출력되어야 합니다.


```python
df = pd.read_csv('data/occupations.csv')
df
```

#### 아래의 방법을 코딩하시오
하지만 우리가 원하는 'gender'에 대한 값은 여기에 나오지 않네요.


'gender'가 'M', 'F' 등의 문자로 되어 있어서, 평균을 구할 수 없기 때문입니다.


그럼 어떻게 해야 평균 성별을 구할 수 있을까요?


남성을 0, 여성을 1로 변경해주면, 숫자로 계산을 할 수 있습니다.

```python
df.loc[df['gender'] == 'M', 'gender'] = 1
df.loc[df['gender'] == 'F', 'gender'] = 0
```

```python
ocu_groups = df.groupby('occupation')
ocu_groups.mean()['gender'].sort_values(ascending=False) # age에 대해서만 계산 될 것
# 그래서 F, M과 같은 명목형을 0과 1로 바꿔줘야 한다.
# 남자를 0, 여자를 1로 했을 때 둘의 평균을 구하면 1이 가까울수록 여자가 많다는 의미.
# 평균이 0.25면 25%가 여자이고, 75%가 남자라는 의미(0쪽에 가까우니)
```

# 11. 데이터 합치기

### pandas _ merge

1. inner join - 컬럼이 겹치는 부분에 대해서만 합치겠다.
2. Left outer join - 왼쪽 데이터 프레임 컬럼에 대해서만 합치겠다. 오른쪽에 없는 것은 NaN이 뜬다.
3. Right outer join - 위와 반대되는 개념.
4. Full outer join - 양쪽 모두

```python
price_df = pd.read_csv('data/vegetable_price.csv') # 데이터프레임 1
quantity_df = pd.read_csv('data/vegetable_quantity.csv') # 데이터프레임 2
```

```python
pd.merge(price_df, quantity_df, on='Product') # inner join - 겹치는 것들만.
```

```python
pd.merge(price_df, quantity_df, on='Product', how='left') # left outer join
```

```python
pd.merge(price_df, quantity_df, on='Product', how='right') # right outer join
```

```python
pd.merge(price_df, quantity_df, on='Product', how='outer') # full outer join
```
