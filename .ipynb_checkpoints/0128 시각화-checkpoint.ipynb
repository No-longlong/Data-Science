### 문자에 대해 선그래프를 그린다면 오류가 발생할 수 있다.

```python
df = pd.read_csv('data/sports.csv', index_col=0)
df.plot(kind='bar') # 세로 그래프
df.plot(kind='barh') # 가로 그래프 horizon
df.plot(kind='bar', stacked=True) # stack은 쌓는다는 의미
```

### 여성에 대한 데이터만 보고 싶다면?

```python
df['Female'] # plot은 Series에서도 가능.
```

```python
df['Female'].plot(kind='bar') # 여성에 대한 bar만 표현
```

# 6. 실리콘 밸리에는 누가 일할까?

```python
df = pd.read_csv('data/silicon_valley_summary.csv')
```

```python
df.head()
```

```python
condition1 = df['job_category'] == 'Managers'
condition2 = df['gender'] == 'Male'
condition3 = df['race_ethnicity'] != 'All'
df.loc[condition1 & condition2 & condition3, :].plot(kind='bar', x='race_ethnicity', y='count')
```

# 9. 실리콘 밸리

```python
df = pd.read_csv('data/silicon_valley_details.csv')
```

```python
adobe = df['company'] == 'Adobe'
race = df['race'] == 'Overall_totals'
count = df['count'] != 0
df_adobe = df[adobe & race & count]
```

```python
df_adobe.set_index('job_category', inplace=True)
```

```python
df_adobe.plot(kind='pie', y='count')
```
