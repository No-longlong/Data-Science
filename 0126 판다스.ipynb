# DataFrame을 만드는 방법이 3가지가 있다.
1. list 그 자체로
2. list를 numpy array에 담은 그 자체로
3. pandas의 Series를 사용한 그 자체로


```python
import numpy as np
import pandas as pd

two_dimensional_list = [['dongwook', 50, 86], ['sineui', 89, 31], ['ikjoong', 68, 91], ['yoonsoo', 88, 75]]
two_dimensional_array = np.array(two_dimensional_list)
list_of_series = [
    pd.Series(['dongwook', 50, 86]), 
    pd.Series(['sineui', 89, 31]), 
    pd.Series(['ikjoong', 68, 91]), 
    pd.Series(['yoonsoo', 88, 75])
]

# 아래 셋은 모두 동일합니다
df1 = pd.DataFrame(two_dimensional_list) # 리스트 자체를 pd.DataFrame에 집어넣은 경우
df2 = pd.DataFrame(two_dimensional_array) # 배열 자체를 집어넣음
df3 = pd.DataFrame(list_of_series) # 

print(df1)
print(df2)
print(df3)
```

# 행, 렬에 있는 값을 찾는 방법?
1. df.loc[행, 렬] --> loc를 쓴 방법
2. df[행] 혹은 df[열] --> loc을 쓰지 않은 깔끔한 방법

# 여러 열의 데이터를 찾으려면?
df[열] 에서 진화해서 df[[열1, 열2]] 로 리스트로 묶어주면 된다.
여튼 [열] 부분은 하나의 파라미터를 반환해야 되는데, 그 파라미터가 리스트라고 생각하면 될듯.

# DataFrame + Dictionary
Dictionary로 구성한 내용을 pd.DataFrame의 파라미터로 하면 동일한 표 형식으로 나타낼 수 있다.

```python
import pandas as pd

samsong_df = pd.read_csv('data/samsong.csv')
hyundee_df = pd.read_csv('data/hyundee.csv')

day = samsong_df['요일']
samsong = samsong_df['문화생활비']
hyundee = hyundee_df['문화생활비']

df_dict = {
    'day' : day,
    'samsong' : samsong,
    'hyundee' : hyundee
}

df = pd.DataFrame(df_dict) # df_dict는 딕셔너리임. 이는 DataFrame을 만드는 파라미터가 될 수 있다.
# 물론 위의 괄호 내의 내용을 df_dict의 값으로 직접 길게 넣어도 된다.

df
```

```python
iphone_df = pd.read_csv('data/iphone.csv', index_col=0)
iphone_df
```

#### row를 여러개 받아오는 방법

```python
#iphone_df.loc[['iPhone 8', 'iPhone X']]

# 연속적인 행에서 슬라이싱으로도 가능하다.
iphone_df.loc['iPhone 8':'iPhone X']
# 이 때에는 리스트 안의 리스트가 아니라는 점을 주의
```

#### column을 여러개 받아오는 방법

```python
iphone_df[['출시일', '디스플레이']]

# 연속적인 행에서 슬라이싱으로도 가능하다.
# 그러나 row와 달리 column에서는 조금 다른 방법
```

```python
# iphone_df['출시일':'디스플레이'] 가 아니다.
iphone_df.loc[:, '출시일':'Face ID']
```

```python
# 위의 경우를 응용해서 원하는 row부터 원하는 column의 범위만 가져오는 방법
iphone_df.loc['iPhone 8':'iPhone X', '출시일':'Face ID']
```

# Pandas에서 Boolean으로 가져오기
df.loc 안의 행 렬을 리스트로 True나 False로 나타내면 해당 인덱스의 값을 가져온다.

True로 입력하지 않은 모든 값은 Default값으로 False로 처리한다.

이 값들은 직접 입력하기보다는 값 > 5 식으로 해당 조건이 Boolean형식으로 나타나는 값을 활용.


**조건문을 사용할 때는 row 이름의 리스트 슬라이싱과 같이 작동합니다.     
즉 iphone_df[iphone_df['디스플레이'] > 5] 과 iphone_df.loc[iphone_df['디스플레이'] > 5]은 같습니다.**

한마디로, [ ] 안의 내용이 bool이면, 해당되는 row만 나타난다.
