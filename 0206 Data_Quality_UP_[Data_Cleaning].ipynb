## 01. 데이터 클리닝: 완결성

```python
import pandas as pd

df = pd.read_csv('data/attendance.csv')
df.head()
# 종목별 평균 관측 수
```

```python
df.isnull()
# 결측값 위치에만 True를 보여준다.
# 그래도 데이터가 크면 눈에 안보인다.
```

```python
df.isnull().sum()
# 각 컬럼별 결측값이 얼마나 있는지를 모두 더해서 보여준다
```

결측값을 해결하는 가장 좋은 방법은 빈칸을 채워넣는 것이다.

그런데, 불가능하다면? 

- 그대로 둔다
- 결측값 레코드를 모두 지워버린다. 

```python
# 레코드 지우기
df.dropna()
# 2010, 2012, 2013년 없어짐 --> 기존데이터 프레임은 건들지 않음
# inplace=True 를 써주면 가능
# 그러나 drop.na()는 나머지 갖고 있는 값들을 모두 잃기 때문에 안좋을 수 있다.
```

```python
df.dropna(axis='columns') # 결측값이 있는 column을 모두 제거
# 배구 관련 데이터 분석이 불가능해진다.
```

있는 값들은 그대로 보존시켜주고싶다면?

```python
df.fillna (0) # 결측값이 있던 부분을 모두 0으로 변경해주기.
```

가장 많이 대체하는 방법은 그 컬럼의 평균값으로 대체해준다.

```python
df.fillna(df.mean()) # 배구의 결측값을 평균값으로 처리
```

```python
df.fillna(df.median()) # 배구의 결측값을 중간값으로 처리 
```

기존 데이터 프레임을 바꾸고 싶으면, inplace=True를 하면 된다.

## 04. 데이터 클리닝: 유일성

```python
df = pd.read_csv('data/dust.csv', index_col=0)
df.head()
```

1. row가 중복
2. column이 중복

- row가 중복

```python
df.index # row인 날짜 컬럼 모두 뽑기 
```

```python
df.index.value_counts() # 날짜 컬럼들 값이 몇번 나오는지.

# 7월 31일만 두번 나온다.
```

```python
df.loc['07월 31일'] # 중복된 row
```

```python
df.drop_duplicates() # 중복된 row를 삭제시키는 메소드
# 기존 데이터프레임을 건들지 않기 때문에 inplace를 해줘야 한다.
```

- column이 중복

```python
# column에서 강원이라는 열이 중복된다.
```

```python
df.T # row와 column을 바꿔준다. 전치행렬 개념
```

```python
df = df.T.drop_duplicates().T
# 아까의 열을 행으로 바꿔주고 중복된 값을 제거시켜준 뒤 이를 다시 전치시켜줌
# 이것은 inplace를 할 수 없기 때문에 (T가 마지막에 썼으므로) 다시 df에 할당시킨다.
```

## 05. 데이터 클리닝: 정확성

### boxplot에 대한 설명
min과 max값은 각각 Q1과 Q3로부터 1.5 * IQR 값만큼 떨어짐을 의미하고

이를 초과하였을시에 이상점으로 판단한다.

```python
%matplotlib inline
```

```python
df = pd.read_csv('data/beer.csv', index_col=0)
df.head()
```

```python
df.plot(kind='box', y='abv')
# 이상점이 3개가 나온다. 
# 데이터프레임에서 지워보자.
# Q1(25%), Q3(75%) 지점을 알고싶다면?
```

```python
# 분위수 보는 방법1
df['abv'].describe()
```

```python
# 분위수 보는 방법2 (파라미터는 분위수 입력)
q1 = df['abv'].quantile(0.25)
q3 = df['abv'].quantile(0.75)
iqr = q3 - q1
```

```python
condition = (df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 * iqr)
# 이상점을 구하는 식(가운데는 or로 나뉜 것)
condition
```

```python
df[condition]
# True인 값이 있는 행만 나타낸다. (이상점 3개)
```

위의 이상점에 대한 설명

첫번째 행은 참이슬 후레시인데 이 데이터들은 맥주 데이터이다. 잘못 들어갔다. 소주는 맥주보다 도수가 높으니 이상점이 된 것.

두번째 행은 보드카가 들어갔기 때문에 도수가 높다.

세번째 행은 맥주가 맞는데 도수가 잘못 입력되었다 5.5%여서 0.055 여야된다.

```python
# 마지막행의 도수만 고치기
df.loc[2250, 'abv'] = 0.055
df.loc[2250]
```

```python
# 맥주가 아닌 두 행은 아예 지워버린다.
df[condition].index # 맥주가 아닌 두 행의 인덱스값을 리스트로 추출
df.drop(df[condition].index, inplace=True)
```

위에서 drop의 파라미터로 df[condition]이 아니라, **.index**를 할당하는 것이 중요하다.

저 자리는 보통 index_col 의 숫자를 써야 되는데, 그 대신 boolean값 중 True의 index를 쓰는 것이다. 

df.drop(2250) 혹은 df.drop(boolean.index)를 해야됨. df.drop(boolean)을 해버리면 에러가 떠버림

그렇기 때문에 단순히 True False로 나타난 Series가 아닌 index값을 입력해줘야 한다.

```python
# 박스플롯을 다시 그려보자
df.plot(kind='box', y='abv')
```

## 07. 데이터 클리닝: 정확성

키가 188cm이고 몸무게가 42kg인 사람은 조금 이상하다

- 관계적 이상점

```python
df = pd.read_csv('data/exam_outlier.csv') # 인덱싱을 위해 인덱스 컬럼을 설정 안함
df.head()
```

```python
df.plot(kind='scatter', x='reading score', y='writing score')
# 위의 데이터를 읽기와 쓰기의 산점도를 그려보면 쓰기를 700점 맞은 이상점을 볼 수 있다.
```

```python
df.corr() # 연속형 변수들간의 상관관계 
# 생각보다 ㅅ읽기와 쓰기의 상관관계가 낮다. 이상점의 영향이 크다고 볼 수 있다.
```

```python
df[df['writing score'] > 100] # 쓰기점수가 680점인 이상점을 발견
```

```python
# 이상점 제거
df.drop(51, inplace=True)
```

```python
df.plot(kind='scatter', x='reading score', y='writing score')
```

```python
# 다시 상관계수를 보면 많이 개선됨을 볼 수 있다.
df.corr() # 0.58에서 0.94로 개선
```

```python
condition = (df['writing score'] > 90) & (df['reading score'] < 40) # 대략 점수 추청해서 조건식 만들기
df[condition]
# 읽기 점수가 매우 낮았다. 분석에 방해가 될 수 있으니 지워야 한다.
```

```python
df.drop(373, inplace=True)
```

```python
df.plot(kind='scatter', x='reading score', y='writing score') # 산점도도 상당히 개선 
df.corr() # 0.94 에서 0.95로 개선
```
