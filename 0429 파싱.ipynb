```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://workey.codeit.kr/music/index')
print(response.text)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
li_tags = soup.select('.popular__order li')
```

```python
li_tags[0].text
```

```python
popular_artists = []
for li in li_tags:
    popular_artists.append(li.text.strip())
```

```python
popular_artists
```

```python
# 모든 전화번호 모으기
response = requests.get("https://workey.codeit.kr/orangebottle/index")
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
phoneNum_tags = soup.select('span.phoneNum')
phoneNum_tags[0].text
```

```python
number = []
for num in phoneNum_tags:
    number.append(num.text)
print(number)
```

```python
# 음악 사이트의 검색어 순위 받아오기
response = requests.get("https://workey.codeit.kr/music/index")
soup = BeautifulSoup(response.text, 'html.parser')
rank_tags = soup.select(".rank__order li")

search_ranks = []
for rank in rank_tags:
    search_ranks.append(rank.text.strip().split(" ")[2])


print(search_ranks)
    
```
