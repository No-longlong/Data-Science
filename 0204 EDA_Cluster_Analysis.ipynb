# 10. Titanic EDA

RMS 타이타닉은 1912년에 빙산과 충돌해 침몰한 여객선입니다. 타이타닉호의 침몰은 무려 1514명 정도가 사망한 비운의 사건이죠. 영화 ‘타이타닉’으로 인해 이름이 가장 널리 알려진 여객선이기도 합니다.


우리에게 주어진 titanic.csv 파일(다운로드)에는 당시 탑승자들의 정보가 담겨 있습니다. 생존 여부, 성별, 나이, 지불한 요금, 좌석 등급 등의 정보가 있는데요.


생존 여부는 'Survived' column에 저장되어 있습니다. 0이 사망, 1이 생존을 의미합니다.
좌석 등급은 'Pclass' column에 저장되어 있습니다. 1은 1등실, 2는 2등실, 3은 3등실을 의미합니다.
지불한 요금은 'Fare' column에 저장되어 있습니다.

다양한 방면으로 EDA(탐색적 데이터 분석)를 한 후, 다음 보기 중 맞는 것을 모두 고르세요

```python
df = pd.read_csv('data/titanic.csv')
df.head()
```

1. 타이타닉 승객은 30대와 40대가 가장 많다.

```python
# 내 풀이
df['Age'].plot(kind='hist')

# 20대와 30대가 가장 많다
```

```python
# 다른 풀이
df.plot(kind='hist', y='Age', bins=30)
# 구체적으로 나누기 위해 bins를 활용
```

*2. 가장 높은 요금을 낸 사람은 30대이다.*

```python
# 내 풀이(한계) - 해결
df.loc[df['Fare'] == df['Fare'].max(), 'Age']
```

```python
# 다른 풀이
df.plot(kind='scatter', x='Age', y='Fare')
# 저 위에 정확한 값의 Age를 알고 싶다면?
```

3. 생존자가 사망자보다 더 많다.
**거짓**

```python
# 내 풀이
df['Survived'].value_counts().plot(kind='pie')
# 0이 사망자, 1이 생존자를 의미한다. 
# 사망자가 더 높은 비율이다.
```

```python
# 다른 풀이
df['Survived'].value_counts()
# 생각해보니 굳이 시각화 할 필요 없다.
```

4. 1등실, 2등실, 3등실 중 가장 많은 사람이 탑승한 곳은 3등실이다.**사실**

```python
# 내 풀이(다른 풀이 동일)
df['Pclass'].value_counts().plot.bar()
# Pclass 중 3의 비율이 가장 높은 것을 볼 수 있다.
# 시각화가 필요 없었다.
```

*5. 가장 생존율이 높은 객실 등급은 1등실이다.*

```python
# 실패
# 다른 풀이
sns.kdeplot(df['Pclass'], df['Survived'])
# 밀집된 쪽이 해당 Pclass의 생존율이 더 높다는 것을 알 수 있다.
```

```python
df['Survived'].plot(kind='pie')
```

6. 나이가 어릴수록 생존율이 높다.

```python
sns.catplot(data=df, x='Survived', y='Age', kind='box')
```

```python
sns.catplot(data=df, x='Survived', y='Age', kind='swarm')
# 대체로 낮게 형성되어 있긴 하나 정확히 모르겠다.
```

```python
# 다른 방법1
sns.stripplot(data=df, x="Survived", y="Age")
```

```python
# 다른 방법2
sns.violinplot(data=df, x="Survived", y="Age")
```

*7. 나이보다 성별이 생존율에 더 많은 영향을 미친다.*

```python
# 다른 방법
sns.stripplot(data=df, x="Survived", y="Age", hue="Sex")
```
