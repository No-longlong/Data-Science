# 01. 새로운 값 계산하기

```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('data/broadcast.csv', index_col=0)
df
```

- 방송사들의 시청률 그래프

```python
df.plot()
```

```python
# 매년 방송사들의 시청률을 연도별 합산한 값
df.sum(axis='columns')
```

```python
df['Total'] = df.sum(axis='columns')
# 새로운 컬럼 'Total' 생성하여 sum한 정보들을 넣어줌
```

```python
df.plot(y='Total')
# Total이라는 컬럼을 플롯으로 나타냄
# 전체적으로 TV의 시청률이 떨어짐을 볼 수 있다.
# 유튜브나 넷플릭스의 영향이라고 볼 수도..
```

```python
df['Group 1'] = df.loc[:, 'KBS':'SBS'].sum(axis='columns')
df['Group 2'] = df.loc[:, 'TV CHOSUN':'MBN'].sum(axis='columns')
```

```python
df.plot(y=['Group 1', 'Group 2'])
# 전체 TV 시청률은 줄어들고 있는데, 지상파를 제외하고 오히려 오르고 있다.
# 지상파 외 채널에서도 많은 프로그램이 탄생하고 있기 때문이다.
```

# 02. 문자열 필터링

```python
df = pd.read_csv('data/albums.csv', encoding='latin1')
df.head(20)
```

### 장르가 Blues가 들어간 것을 찾고 싶다.
#### 그러나 우리가 이전에 배운 개념은 정확히 Blues인것만 찾을 수 있다.

```python
df[df['Genre'] == 'Blues'] # 이전에 배웠던 개념. 정확히 Blues만 찾을 수 있다.
```

#### 여러장르 중에서 Blues라는 문자가 섞인 장르를 찾고자 할때는?

```python
df[df['Genre'].str.contains('Blues')] # 장르들 중에서 Blues라는 문자가 포함된 것들만 나타냄
```

#### 여러장르 중에서 Blues라는 문자로 시작하는 장르를 찾고자 할때는?

```python
# Blues로 시작하는 값만 골라내기
df[df['Genre'].str.startswith('Blues')]
```

```python
df['Contains Blues'] = df['Genre'].str.contains('Blues')
df
# Blues라는 문자만 포함된 장르만을 불린값으로 만들어서 새로운 컬럼 만들기
```
