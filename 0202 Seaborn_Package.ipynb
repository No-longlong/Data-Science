# 4. KDE Plot

```python
import pandas as pd
import seaborn as sns
body_df = pd.read_csv('data/body.csv', index_col=0)
body_df.head()
```

```python
body_df['Height']
```

```python
# 카운트를 파악
body_df['Height'].value_counts().sort_index()
# 인덱스로 정렬
```

```python
body_df['Height'].value_counts().sort_index().plot()
```

### 아래는 우리의 데이터셋을 가지고 KDE를 추측한 것일뿐, 실제 분포와 차이가 있다.

```python
sns.kdeplot(body_df['Height'])
# sns의 kdeplot에 body_df의 키높이를 넘겨준것. 아래의 넓이는 1이다.
# 파라미터로 bw=0.5 처럼 몇개의 구간으로 나누는지의 개념과 비슷
```

# KDE 활용 예시

```python
# box plot은 통계적으로, violinplot은 KDE plot을 대칭형으로 나타낸 것처럼, 이는 원래 바이올린 모양이라해서 이름이 저렇게 된 것.
```

```python
sns.violinplot(body_df['Height'])
```

```python
sns.kdeplot(body_df['Height'], body_df['Weight'])
# 등고선처럼 높낮이를 표현, 파라미터 2개를 넣는 것은 x, y 를 의미한다.
```

등고선 설명을 위해 아래와 같이 x축 y축을 비교해보자.

x축에는 키에대한 kde, y축은 몸무게 kde를 3차원으로 만든 것

선이 가까울수록 가파른거고, 선끼리 멀수록 평평한 것 (x축, y축 방향으로 이해해야 한다)

```python
sns.kdeplot(body_df['Height'])
```

```python
sns.kdeplot(body_df['Weight'])
```
